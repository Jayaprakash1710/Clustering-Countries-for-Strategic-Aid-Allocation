# Univariate Analysis:

```python
numerical_columns=list(df.columns)[1:]
numerical_columns
```

```python
plt.figure(figsize=(16,20))
for i in range(1,len(numerical_columns)+1):
  plt.subplot(5,2,i)
  ax=sns.histplot(df[numerical_columns[i-1]],kde=True,line_kws={'linewidth':3})
  ax.lines[0].set_color('red')
plt.show()
```
